```python
import pandas as pd
```

```python
# Завантажую файли
df_2016 = pd.read_csv(r'C:\Users\PC\Google_trends_2016-2017.csv')
df_2026 = pd.read_csv(r'C:\Users\PC\Google_trends_2026-2025.csv')
df_2016
```

```python
# 'Скіпаю' перший рядок щоб назви колонок були правильні
df_2016 = pd.read_csv(r'C:\Users\PC\Google_trends_2016-2017.csv', skiprows=2)
df_2026 = pd.read_csv(r'C:\Users\PC\Google_trends_2026-2025.csv', skiprows=2)
df_2016
```

```python
# Перейменовую рядки для подальшої роботи в sql
cols = ['date', 'bukovel', 'yaremche' ]
df_2016.columns = cols
df_2026.columns = cols
df_2016
```

```python
df_2016.dtypes
```

```python
# Встановлюю тип даних для колонки 'date'
df_2016['date'] = pd.to_datetime(df_2016['date'])
df_2026['date'] = pd.to_datetime(df_2026['date'])
```

```python
df_2016.dtypes
```

```python
# Прибираю рядок з датою яка не потрібна
df_2016 = df_2016.iloc[1:].copy()
df_2026 = df_2026.iloc[1:].copy()
df_2016
```

```python
df_2026
```

```python
# Створююнову колонку для подальшого об'єднання даних
df_2016['period'] = 'Old'
df_2026['period'] = 'New'
df_2016.head()
```

```python
# Об'єдную дані в одну базу
df_final = pd.concat([df_2016, df_2026])
df_final
```

```python
# Зберігаю остаточний файл
df_final.to_csv('cleaned_data_Ukraine.csv', index=False)
```
